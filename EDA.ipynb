```python
import numpy
import matplotlib.pyplot as plt
import cv2
import glob
import random
```

```python
path = ''
```

```python
imgs = []
counts = []
labels = []

for folder_path in glob.glob(path + 'Training/Training/*'):
    labels.append(folder_path.split('\\')[-1])
    counts.append(len(glob.glob(folder_path + '/*.*')))
    img_path =  random.choice(glob.glob(folder_path + '/*.*'))
    imgs.append(cv2.imread(img_path))
```

```python
plt.figure()

fig, axes = plt.subplots(1, 2, figsize = (16, 16))

for i, img in enumerate(imgs):

    plt.subplot(1, 2, i + 1)
    
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(f'{labels[i]}: {counts[i]}')

plt.show()
```

```python
fire_paths = []
for img_path in glob.glob(path + 'Training/Training/Fire/*.*'):
    fire_paths.append(img_path)
```

```python
fire_imgs = []
for img_path in random.sample(fire_paths, 9):
    fire_imgs.append(cv2.imread(img_path))
```

```python
plt.figure()

fig, axes = plt.subplots(3, 3, figsize = (16, 16))

for i, img in enumerate(fire_imgs):

    plt.subplot(3, 3, i + 1)
    
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(f'{img.shape}')

plt.show()
```
